# Gaussian Updated Fields + analytics for magnetic dipole

```python
import sys
sys.path.append('..')
```

```python
# Loading my modules
from plotting_tools.plotter import plotter
from plotting_tools.field_comparison import differencer
from plotting_tools.sample_field import sample_field
from physics_informed_vector_fields.magnetic_dipole import field_around_magnetic_diplole
from GP_Implementation.Kernel import updated_vector_field
```

```python
# getting the fields to draw samples from
outputs_dipole = field_around_magnetic_diplole(100000000, masking=0.25)
```

```python
sampled_outputs_dipole = sample_field(outputs_dipole, n_samples=1000)
plotter(sampled_outputs_dipole, 
        title = 'Points given to our GP Regressor for magnetic dipole')
```

## Applying GP Regression to magnetic dipole

```python
given_samples_dipole = sampled_outputs_dipole[0].shape[0]

outputs_dipole_div_free = updated_vector_field(
    sampled_outputs_dipole, 
    outputs_dipole[0], outputs_dipole[1], outputs_dipole[2], 
    kernel_func='divergence_free_kernel', 
    sigma_f = 0.01, l = 1)
```

```python
plotter(outputs_dipole_div_free, 
        title = f'Div free output from GP Regression for n = {given_samples_dipole} input vectors for magnetic dipole')
```

```python
# plot residuals of fit above: 
residuals_dipole_div_free, mse_dipole_div_free = differencer(outputs_dipole, outputs_dipole_div_free)
print(f'MSE for this case of {mse_dipole_div_free}')
plotter(residuals_dipole_div_free, 
        title = f'Div free regression errors for n = {given_samples_dipole} input vectors for magnetic dipole')
```

```python
outputs_dipole_curl_free= updated_vector_field(
    sampled_outputs_dipole, 
    outputs_dipole[0], outputs_dipole[1], outputs_dipole[2], 
    kernel_func='curl_free_kernel',
    sigma_f = 0.1, l = 0.3)
```

```python
plotter(outputs_dipole_curl_free, 
        title = f'Curl free output from GP Regression for n = {given_samples_dipole} input vectors for current carrying dipole')
```

```python
# plot residuals of fit above: 
residuals_dipole_curl_free, mse_dipole_curl_free = differencer(outputs_dipole, outputs_dipole_curl_free)
print(f'MSE for this case of {mse_dipole_curl_free}')
plotter(residuals_dipole_curl_free, 
        title = f'Curl free regression errors for n = {given_samples_dipole} input vectors for magnetic dipole')
```
